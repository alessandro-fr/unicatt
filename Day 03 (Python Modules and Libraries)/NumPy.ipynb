```python
my_array = np.array([1,10,-5,20])
my_array
```

```python
arr1 = np.random.randint(0,10,4)
arr2 = np.random.randint(0,10,4)

print(f"The two arrays are: \n{arr1} \n{arr2}")
```

```python
print(f"The sum of the two arrays is {arr1 + arr2}")
print(f"The difference of the two arrays is {arr1 - arr2}")
print(f"The multiplication of the two arrays is {arr1 * arr2}")
```
