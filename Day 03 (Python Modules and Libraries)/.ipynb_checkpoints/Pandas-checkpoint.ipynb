```python
df2 = df2[df2["total (R$)"]<1e4]
df2
```

```python
max_price = df2["total (R$)"].max()
print(f"Maximum price of the houses is {max_price}")
```

```python
threashold = max_price * 0.9
print(f"The threashold is {threashold}")
```

```python
top_df = df2[df2["total (R$)"]>threashold]
top_df
```

```python
top_df_sorted = top_df.sort_values(
    by = "total (R$)",
    ascending = False)

top_df_sorted.head (20)
```
