```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,-50,100,11]
```

# Bar Chart

```python
categories = ["Company A", "Company B", "Company C", "Company D"]

values = [20,30,35,15]

colors = ["red", "green", "blue", "orange"]

plt.bar(categories, values, color = colors)
plt.title ("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.xticks(rotation=45)

plt.show()
```

```python
import mpl_toolkits.mplot3d as Axes3D


categories = ["Company A", "Company B", "Company C", "Company D"]
values = [20, 30, 35, 15]
colors = ["red", "green", "blue", "orange"]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# coordinate X
x_pos = range(len(categories))
y_pos = [0]*len(categories)   # tutte le barre sulla stessa linea Y
z_pos = [0]*len(categories)

# dimensioni barre
dx = [0.5]*len(categories)
dy = [0.5]*len(categories)
dz = values

ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, color=colors)

ax.set_xticks(x_pos)
ax.set_xticklabels(categories, rotation=45)
ax.set_ylabel("Y")
ax.set_zlabel("Values")
ax.set_title("3D Bar Chart")

plt.show()
```
