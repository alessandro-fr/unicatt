# While loops

```python
counter =0
while counter <5:
    print(f'within the loop: {counter}')
    counter += 1
    if counter == 3:
        break

print(f'outside of the loop: {counter}')
```

```python
counter = 10
while counter >= 1:
    print(counter)
    counter /= 2
    if counter < 1:
        break
print(counter)
```

```python
for i in range(2, 21, 2):
    print(i)
```

```python
for i in range(3): # loop for three rows
    for j in range(5): #loop for five columns
        print('*', end='')  # end for putting * on the same line
    print()
```

# Error handling

```python
try:
    x = 3/ 0
except ZeroDivisionError as e:
    print(e)
```

```python
try:
    x = 10 + 'abc'
except TypeError as e:
    print(f'The following error occurred: {e}')
```

```python
try:
    my_list = [1,2,3,4,5]
    print(my_list[10])
except Exception as e:
    print(f'The following error type has happened: {e}')
```
