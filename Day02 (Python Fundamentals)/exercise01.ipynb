```python
def is_even_2(a):
    """
    this function returns True if the number is even,
    otherwise it returns False
    """
    if a%2 == 1:
            return False
    else:
        return True

print(is_even_2(11))

print(is_even_2(10))
```

```python
def is_even_3(a):
    """
    this function returns True if the number is even,
    otherwise it returns False
    """

    if a%2 == 0:
            return True
    
    return False

print(is_even_3(11))

print(is_even_3(10))
```

```python
def is_even_4(a):
    """
    this function returns True if the number is even,
    otherwise it returns False
    """

    return (a%2 == 0) # return True if the result is 0, otherwise returns False

print(is_even_3(11))

print(is_even_3(10))
```
